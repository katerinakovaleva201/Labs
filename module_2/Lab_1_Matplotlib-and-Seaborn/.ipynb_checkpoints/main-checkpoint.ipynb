#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

import math
x = np.arange(0,100)
y = x**2
z = math.e**x
```

```python
fig, ax = plt.subplots(1,2,figsize = (6,3))

ax1 = ax.flatten()[0]
ax1.plot(x,y, label='y=x*2')
ax1.plot(x,z, label='y=x**2')
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2 = ax.flatten()[1]
ax2.plot(x,y, label='y=x*2')
ax2.plot(x,z, label='y=x**2')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel("x")
ax2.set_ylabel("y")
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
avg_steps=fitbit.groupby(['Months_encoded','Work_or_Weekend'])[['Steps']].mean('Steps').reset_index()
avg_steps.columns
```

```python
median_steps=fitbit.groupby('Months_encoded')[['Steps']].median('Steps').reset_index()
```

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
plt.title('Monthly steps on weekdays or weekend', size='20',)

ax=sns.barplot(x = avg_steps['Months_encoded'], y = avg_steps['Steps'], hue = avg_steps['Work_or_Weekend'])

plt.sca(ax)
plt.plot(median_steps,  color='c', label='median steps')

plt.legend()
# plt.savefig("figure1.jpg")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

y_labels = ['Minutes Sedentary', 'Minutes Lightly Active','Minutes Very Active']
colors = ['green','blue','orange']
fig5, axes5 = plt.subplots(1,3, figsize = (12,4))
plt.sca(axes5[0])
for i in range(0,3):
    sns.scatterplot(x = fitbit[y_labels[i]], y=fitbit['Steps'], ax = axes5[i], color = colors[i])
```
