```python
import pandas as pd
```

```python
#Upload the data

df = pd.read_excel("example_data_cleaning.xlsx")
df.head()
```

```python
# Identify and fill missing values

null_columns = df.isnull().sum()/df.size
null_columns
```

```python
df['BirthYear']= df['BirthYear'].fillna(0)
df['BirthYear'] = df['BirthYear'].astype('int64')
df
```

```python
df['Profession']= df['Profession'].fillna("Not Specified")
df
```

```python
# Examining Data for Potential Issues

df.columns = df.columns.str.rstrip(' ')
df.head()
```

```python
print(df['Profession'].unique())
print('\n')
print(df['Risk'].unique())
```

```python
df['Profession'] = df['Profession'].str.replace('etudient', 'Student')
df['Profession'] = df['Profession'].str.capitalize()
df['Profession'] = df['Profession'].str.replace('Hr', 'HR')
df['Risk'] = df['Risk'].fillna('Not specified')
df
```

```python
# Search for duplicates
len(df)
```

```python
df = df.drop_duplicates()
len(df)
# There are no duplicates
```

```python
# Encode categorical data

pd.get_dummies(df)
```
